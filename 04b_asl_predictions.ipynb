```python
from tensorflow import keras

model = keras.models.load_model('asl_model.h5')
```

```python
model.summary()
```

```python
image_path=r'C:\Users\Dell\OneDrive\Desktop\Project\asl_alphabet_test\asl_alphabet_test\Q_test.jpg'
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def show_image(image_path):
    image = mpimg.imread(image_path)
    plt.imshow(image, cmap='gray')
```

```python
show_image(image_path)
```

```python
from tensorflow.keras.preprocessing import image as image_utils

def load_and_scale_image(image_path):
    image = image_utils.load_img(image_path, color_mode="grayscale", target_size=(28,28))
    return image
```

```python
image = load_and_scale_image(image_path)
plt.imshow(image, cmap='gray')
```

```python
image = image_utils.img_to_array(image)
```

```python
image = image.reshape(1,28,28,1) 
```

```python
image = image / 255
```

```python
prediction = model.predict(image)
print(prediction)
```

```python
import numpy as np
np.argmax(prediction)
```

```python
alphabet = "abcdefghiklmnopqrstuvwxy"
```

```python
alphabet[np.argmax(prediction)]
```

```python
def predict_letter(file_path):
    show_image(file_path)
    image = load_and_scale_image(file_path)
    image = image_utils.img_to_array(image)
    image = image.reshape(1,28,28,1) 
    image = image/255
    prediction = model.predict(image)
    # convert prediction to letter
    predicted_letter = alphabet[np.argmax(prediction)]
    return predicted_letter
```

```python
predict_letter(image_path)
```

```python
predict_letter(image_path)
```
